```python
%matplotlib inline
import scipy
from matplotlib import pyplot as plt
import matplotlib as mpl

def plot(n):
    scipy.random.seed(1)
    x = scipy.random.normal(-5, 1, size=n)
    y = scipy.random.normal(-5, 1, size=n)
    c = (((x + 5) ** 2) + ((y + 5) ** 2)) ** 0.5
    fig = plt.figure(figsize = (8,5))
    ax = fig.add_axes([0,0,0.95,1])
    ax1 = fig.add_axes([0.95, 0, 0.05, 1])
    ax.scatter(x, y, s = 20, c = c, cmap=mpl.cm.viridis)
    ax.set_ylabel('Y')
    ax.set_xlabel('X')
    ax.grid(alpha = 0.5)
    ax.set_title('2D Normally Distributed '  + str(n) + ' Points')
    fig.subplots_adjust(right=0.5)
    cmap = mpl.cm.viridis
    norm = mpl.colors.Normalize(vmin=min(c), vmax=max(c))
    cb1 = mpl.colorbar.ColorbarBase(ax1, cmap=cmap, norm=norm)
    cb1.set_label('Distance from mean')
    fig.show()
```

```python
plot(100)
```

```python
plot(1000)
```
